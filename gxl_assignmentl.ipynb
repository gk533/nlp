对文本进行整理，划分词组

````python
#按照词组格式进行划分#
def create_grammar(grammar_str, split='=', line_split='\n'):
    grammar = {}
    for line in grammar_str.split(line_split):
        if not line.strip(): continue
        exp, stmt = line.split(split)
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar
````

````python
create_grammar(host)
````

````python
#按照语法规则随机生成句子
import random
choice = random.choice
def generate(gram, target):
    if target not in gram: return target 
    
    expaned = [generate(gram, t) for t in choice(gram[target])]
    
    return ''.join([e if e != '/n' else '\n' for e in expaned if e != 'null'])
````

````python
def generate_n(n):
    for i in range(n):
        print('host speak:')
        print(generate(gram=create_grammar(host, split='='), target='host'))
        print('human answer:')
        print(generate(gram=create_grammar(human, split='='), target='human'))
````

````python
import pandas as pd
````

````python
content = pd.read_csv('movie_comments.csv',low_memory=False)
````

````python
content.head()
````

````python
comments = content['comment'].tolist() # Serise转换为List
````

````python
comments[:3]
````

````python
import re
def token(string):
    # we will learn the regular expression next course.
    return re.findall('\w+', string)          #去掉句子里的标点符号
````

````python
comments_clean = [''.join(token(str(a))) for a in comments]
````

````python
comments_clean[0:3]
````

````python
import jieba
````

````python
def cut(string): return list(jieba.cut(string))
````

````python
def count_token():
    count=0
    TOKEN=[]
    for i in comments_clean:
        if count % 50 == 0:
            print(count)
        if count > 10000:
            break
        count += 1
        TOKEN += cut(i)
    return TOKEN
````

````python
TOKEN=count_token()
````

````python
from collections import Counter
words_count = Counter(TOKEN)
words_count
````

````python
import matplotlib.pyplot as plt

frequiences = [f for w, f in words_count.most_common(100)]
x = [x for x in range(100)]
import matplotlib.pyplot as plt
plt.plot(x,frequiences)
````

````python
def pro_1(word): # 单个词概率
    return words_count[word] / len(TOKEN)
````

````python
pro_1('吴京')
````

````python
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-1]))]
````

````python
words_count_2 = Counter(TOKEN_2_GRAM)
words_count_2
````

````python
def pro_2(word_1, word_2): #两个词组合概率
    if word_1 + word_2 in TOKEN_2_GRAM:
        return words_count_2[word_1 + word_2] / len(TOKEN_2_GRAM)
    else:
        return 1 / len(TOKEN_2_GRAM)
````

````python
pro_2('吴京', '电影')
````

````python
def get_probablity(sentence):
    words = cut(sentence)
    
    sentence_pro = 1
    
    for i, word in enumerate(words[:-1]):
        next_ = words[i+1]
        
        probability = pro_2(word, next_)
        
        sentence_pro *= probability
    
    return sentence_pro
````

````python
for sen in [generate(gram=create_grammar(human, split='='), target='human') for i in range(10)]:
    print('sentence: {} with Prb: {}'.format(sen, get_probablity(sen)))
````

能够让list按照第1个元素进行排序, 但是是递减的顺序。

````python
def generate_best(gram, model, n): # you code here
    sentences_pro_list = []
    for i in range(n):
        sentence = generate(create_grammar(gram[0]), gram[1])
        sentence_pro = model(sentence)
        sentence_pro_list = (sentence, sentence_pro)
        sentences_pro_list.append(sentence_pro_list)
    print(sentences_pro_list)
    return sorted(sentences_pro_list,key=lambda x:x[1], reverse=True)[0]
````
